```python
from utils.myutils import split_triple, save_object
```

```python
file_path_abstracts = "dbpedia/long_abstracts_en.ttl"

abstracts_file = open(file_path_abstracts, 'r', encoding='utf-8')

abstract_dict = dict()
```

```python
i = 0
for line in abstracts_file:
    i+= 1
        
    if i % 1000 == 0:
        print(str(i) + " entries processed")
        
    if line[:1] == '#' or len(line) == 0:
        continue

    try:
        triple = split_triple(line)
    except ValueError:
        print("ValueError: Split not possible: " + line)
        continue
        
    if(triple[1] != '<http://dbpedia.org/ontology/abstract>'):
        print("Unknown middle! It is " + triple[1])
        continue
    
    abstract_dict[triple[0]] = triple[2]
```

```python
save_object(abstract_dict, "abstract_dict")
```
