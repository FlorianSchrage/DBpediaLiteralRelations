```python
from utils.myutils import load_object, split_triple, save_object
```

```python
file_path_triples_filtered_sorted_by_relation = "dbpedia/mappingbased_literals_en_filtered_sorted_by_relation.ttl"

triples_file_filtered_sorted_by_relation = open(file_path_triples_filtered_sorted_by_relation, 'r', encoding='utf-8')

instance_types_reverse_dict = load_object("instance_types_reverse_dict")
domain_dict = load_object("domain_dict")

candidate_dict = dict()
```

```python
prev = None
entities_already_present = set()
triples_file_filtered_sorted_by_relation.seek(0)
for i, line in enumerate(triples_file_filtered_sorted_by_relation):
    
    triple = split_triple(line)
    
    if triple[1] != prev:
        if prev != None:
            print("---" + prev + "---")
            try:
                domain = domain_dict[prev[1:-1]]
                print("Domain: " + str(domain))
            except KeyError:
                print("KeyError! No domain for " + prev)
            try:
                #Get all objects that fit the domain of the relation
                candidates = instance_types_reverse_dict['<' + domain + '>']
                #Only consider those that do not already exist
                #print("Number of Candidates before: " + str(len(candidates)))
                candidates = [c for c in candidates if not c in entities_already_present]

                #print("Number of Candidates: " + str(len(candidates)))

                candidate_dict[prev] = candidates
            except KeyError:
                print("KeyError! " + domain)

            entities_already_present = set()

        prev = triple[1]
    
    entities_already_present.add(triple[0])
    #print(str(entities_already_present))
    
    if i % 1000 == 0:
        pass
        #print(str(i) + " entries processed")

try:
    domain = domain_dict[prev]
    print("Domain: " + str(domain))
except KeyError:
    print("KeyError! No domain for " + prev)
try:
    candidates = instance_types_reverse_dict['<' + domain + '>']
    candidates = [c for c in candidates if not c in entities_already_present]
    candidate_dict[prev] = candidates
except KeyError:
    print("KeyError! " + domain)
```

```python
save_object(candidate_dict, "candidate_dict")
```

```python
for k, v in candidate_dict.items():
    print(k + ': ' + str(len(v)))
```
