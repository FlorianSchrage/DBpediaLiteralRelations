```python
from utils.myutils import split_triple, save_object
```

```python
file_path_instance_types_transitive = "dbpedia/instance_types_transitive_en.ttl"
file_path_instance_types = "dbpedia/instance_types_en.ttl"

instance_types_transitive_file = open(file_path_instance_types_transitive, 'r', encoding='utf-8')
instance_types_file = open(file_path_instance_types, 'r', encoding='utf-8')

instance_types_dict = dict()
instance_types_list = list()
```

```python
prev = None
for i, line in enumerate(instance_types_transitive_file):        
    if i % 1000 == 0:
        print(str(i) + " entries processed")
        
    if line[:1] == '#' or len(line) == 0:
        continue

    try:
        triple = split_triple(line)
    except ValueError:
        print("ValueError: Split not possible: " + line)
        continue
        
    if(triple[1] != '<http://www.w3.org/1999/02/22-rdf-syntax-ns#type>'):
        print("Unknown middle! It is " + triple[1])
        continue
    
    if prev != None and prev != triple[0]:
        instance_types_dict[prev] = instance_types_list
        instance_types_list = list()
    
    instance_types_list.append(triple[2])
    prev = triple[0]
    
instance_types_dict[prev] = instance_types_list

#Second file:
print('Now processing 2nd file...')
for i, line in enumerate(instance_types_file):
    if i % 1000 == 0:
        print(str(i) + " entries processed")
        
    if line[:1] == '#' or len(line) == 0:
        continue

    try:
        triple = split_triple(line)
    except ValueError:
        print("ValueError: Split not possible: " + line)
        continue
    
    if(triple[1] != '<http://www.w3.org/1999/02/22-rdf-syntax-ns#type>'):
        print("Unknown middle! It is " + triple[1])
        continue
    try:
        if triple[2] not in instance_types_dict[triple[0]]:
            instance_types_dict[triple[0]].append(triple[2])
    except KeyError:
        instance_types_dict[triple[0]] = [triple[2]]

print('Finished')
```

```python
save_object(instance_types_dict, "instance_types_dict")
```

```python
#Invert dict
instance_types_reverse_dict = dict()
for entity, instance_types_list in instance_types_dict.items():
    for instance_type in instance_types_list:
        
        if instance_type in instance_types_reverse_dict:
            instance_types_reverse_dict[instance_type].append(entity)
        else:
            instance_types_reverse_dict[instance_type] = [entity]
```

```python
save_object(instance_types_reverse_dict, "instance_types_reverse_dict")
```
