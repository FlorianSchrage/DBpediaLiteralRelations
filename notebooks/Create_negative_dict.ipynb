```python
from utils.myutils import split_triple, load_object, save_object
```

```python
file_path_triples_filtered_sorted_by_relation = "dbpedia/mappingbased_literals_en_filtered_sorted_by_relation.ttl"

triples_file_filtered_sorted_by_relation = open(file_path_triples_filtered_sorted_by_relation, 'r', encoding='utf-8')

relation_type_dict = load_object('relation_type_dict')
types_date = load_object('types_date')

negative_dict = dict()
```

```python
prev = None
positive_type = False
all_numbers = list()
for i, line in enumerate(triples_file_filtered_sorted_by_relation):
    triple = split_triple(line)
    
    #Skip dates
    if relation_type_dict[triple[1][1:-1]] in types_date:
        continue
    
    if prev != triple[1] and prev != None:
        
        negative_numbers = [n for n in all_numbers if n < 0]
        negative_percentage = len(negative_numbers) / len(all_numbers)
        #Min 0.005 (aus praktischen Versuchen ermittelt)
        if positive_type or negative_percentage < 0.005:
            negative_dict[prev] = False
        else:
            negative_dict[prev] = True
            
        all_numbers = list()
        positive_type = False
        
    
    if 'nonNegative' in triple[2]:
        positive_type = True
    
    fact_number = triple[2].split("^^")[0]
    all_numbers.append(float(fact_number))
    
    if i % 1000 == 0:
        print(str(i) + " entries processed")
    
    prev = triple[1]

egative_numbers = [n for n in all_numbers if n < 0]
negative_percentage = len(negative_numbers) / len(all_numbers)
if positive_type or negative_percentage < 0.005:
    negative_dict[prev] = False
else:
    negative_dict[prev] = True
```

```python
save_object(negative_dict, 'negative_dict')
```
