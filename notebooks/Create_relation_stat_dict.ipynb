```python
import statistics
from utils.myutils import split_triple, save_object, load_object
```

```python
file_path_triples_filtered_sorted_by_relation = "dbpedia/mappingbased_literals_en_filtered_sorted_by_relation.ttl"

triples_file_filtered_sorted_by_relation = open(file_path_triples_filtered_sorted_by_relation, 'r', encoding='utf-8')

relation_type_dict = load_object("relation_type_dict")
types_int = load_object("types_int")
types_float = load_object("types_float")

relation_stat_dict = dict()
```

```python
prev = None
all_numbers = list()
for i, line in enumerate(triples_file_filtered_sorted_by_relation):
    if i % 1000 == 0:
        print(str(i) + ' entires processed.')
        
    triple = split_triple(line)
    
    if relation_type_dict[triple[1][1:-1]] not in types_int and relation_type_dict[triple[1][1:-1]] not in types_float:
        continue
    
    if triple[1] != prev and prev != None:
        mean = statistics.mean(all_numbers) #sum(all_numbers) / float(len(all_numbers))
        stdev = statistics.stdev(all_numbers)
        relation_stat_dict[prev] = (mean, stdev)
        all_numbers = list()
    
    all_numbers.append(float(triple[2].split("^^")[0]))
    prev = triple[1]

mean = statistics.mean(all_numbers) #sum(all_numbers) / float(len(all_numbers))
stdev = statistics.stdev(all_numbers)
relation_stat_dict[prev] = (mean, stdev)
all_numbers = list()
```

```python
save_object(relation_stat_dict, 'relation_stat_dict')
```
