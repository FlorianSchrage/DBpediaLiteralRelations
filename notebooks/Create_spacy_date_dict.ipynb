```python
import spacy
from utils.myutils import split_triple, save_object, load_object
```

```python
abstract_dict = load_object('abstract_dict')

nlp = spacy.load('en')

spacy_date_dict = dict()
```

```python
for i, relation in enumerate(abstract_dict):
        
    if i % 1000 == 0:
        print(str(i) + " entries processed")
        
    doc = nlp(abstract_dict[relation])
    
    dates = [(ent.start_char, ent.end_char) for ent in doc.ents if ent.label_ == 'DATE']
    
    spacy_date_dict[relation] = dates
```

```python
save_object(spacy_date_dict, "spacy_date_dict")
```

```python
for i, relation in enumerate(spacy_date_dict):
    print(relation + " -> " + str(spacy_date_dict[relation]))
```
