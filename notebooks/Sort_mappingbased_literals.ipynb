```python
from utils.myutils import split_triple
from operator import itemgetter
```

```python
#Parameter for the file name
sort_by = "subject"

#Parameter Number for itemgetter
sort_by_number = 0
```

```python
file_path_triples_filtered = "dbpedia/mappingbased_literals_en_filtered.ttl"
file_path_triples_filtered_sorted_by_relation = "dbpedia/mappingbased_literals_en_filtered_sorted_by_" + sort_by + ".ttl"

output = open(file_path_triples_filtered_sorted_by_relation, 'w', encoding='utf-8')
```

```python
with open(file_path_triples_filtered, 'r', encoding='utf-8') as f:
    lines = [split_triple(line) for line in f]

i = 0
for line in sorted(lines, key=itemgetter(sort_by_number)):
    i+= 1
    output.write(' '.join(line) + " .\n")
    
    if i % 1000 == 0:
        print(str(i) + " entries processed")

output.close()
```
